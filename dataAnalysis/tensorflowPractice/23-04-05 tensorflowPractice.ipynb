```python
import pandas as pd

data = pd.read_csv('test.csv')

# print(data.isnull().sum()) 비어있는 값의 개수

data = data.dropna()    #null값 행 삭제

y = data['admit'].values # admit안의 값들 list로 만듦
x = []

for i, rows in data.iterrows():    #data.iterrow(): 데이터 프레임을 한 행씩 출력  i : 행 번호 rows : values
    x.append([rows['gre'],rows['gpa'],rows['rank']]) # list x에 행마다 gre, gpa, rank 의 value를 삽입

#데이터 전처리.
#값이 비어있을 때 1. 행 삭제 (데이터 적을 때만) 2. 임의의 평균값을 채워 넣는다.
#fillna(100) 빈칸들 100으로 채움, data['gpa'] gpa열, data['gpa'].min() and count() gpa열의 최솟값 and 갯수

import numpy as np
import tensorflow as tf

model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation = 'tanh'), # 일반적인 히든레이어 node 64개
    tf.keras.layers.Dense(128, activation = 'tanh'), #sigmoid = 0~1사이의 확률
    tf.keras.layers.Dense(1, activation = 'sigmoid'), # 마지막은 하나의 node가 있어야한다. output Layer.
])

model.compile(optimizer= 'adam', loss= 'binary_crossentropy', metrics=['accuracy']) 
#binary_crossentropy: 구할 값이 0과 1사이의 분류/확률일 때 많이 사용한다.

model.fit(np.array(x), np.array(y), epochs=1000) # w 값 최적화 #파이썬 리스트를 np.array로 변환해야함

# model.fit(x,y,epochs) x = 학습시킬 데이터, y = 실제 정답, epochs = 학습 할 데이터의 양

#loss -> 손실: 학습할수록 낮아져야함, accuracy -> 정답률:  학습할수록 높아져야함

# 예측값
predict = model.predict([[750,3.70,3],[400,2.2,1]])
print(predict)

#성능 향상 -> 데이터 전처리 or 레이어의 추가 및 변경.  연구 및 실험이 중요
```
